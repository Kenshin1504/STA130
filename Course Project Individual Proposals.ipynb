**Research Question #1:** Is there a correlation among Canadians between burnout levels and feelings of loneliness?

**Variables:**

Predictor: WELLNESS_malach_pines_burnout_measure_score

Outcome: LONELY_ucla_loneliness_scale_score

Burnout and loneliness both affect personal well-being and the quality of social connections. Since both WELLNESS_malach_pines_burnout_measure_score and LONELY_ucla_loneliness_scale_score are numerical variables, a scatter plot may be appropriate, allowing for a visual assessment of any correlation and outliers in the data.

**Analysis:**
An Ordinary Least Squares (OLS) simple linear regression model will be used to find a best-fitting line with WELLNESS_malach_pines_burnout_measure_score predicting LONELY_ucla_loneliness_scale_score. This will give us the test statistic for the slope coefficient, which can be used to calculate the p-value for the test statistic. After setting an $\alpha$-significance level, the p-value can be calculated through permutation testing to generate a sampling distribution of slope coefficients under the null hypothesis. We assume linearity, independence of observations as well as independence, homoscedasticity and normality of the errors, which should have a mean of zero. 

**Hypotheses:**
Null hypothesis ($H_0$): There is no linear association between burnout levels and feelings of loneliness    ($β_1 = 0$).

Alternative hypothesis ($H_1$): There is a linear association between  burnout levels and feelings of loneliness ($β_1 \not= 0$).

**Possible Results:**

1. Statistically significant positive association: The p-value is less than the $\alpha$-significance level and the test statistic is positive.
2. Statistically significant negative association: The p-value is less than the $\alpha$-significance level and the test statistic is negative. 
3. No statistically significant association: The p-value is greater than the $\alpha$-significance level.

These results will determine whether a person's burnout level is positively, negatively or not linearly associated with their feelings of loneliness.
