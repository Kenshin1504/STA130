Task 4

df.shape provides the total number of columns in the DataFrame, whereas df.describe() only summarises numeric columns by default and hence excludes columns of a non-numeric type. This leads to df.describe() showing less columns than df.shape for this particular dataset.

df.shape counts all entries for each column, including null entries which imply missing values. On the other hand, the "count" value reported by df.describe() reflects the number of non-null entries only. Any null entries, and hence missing values, are excluded. Thus, the values which df.describe reports in the "count" column are equal to the total count, as reported by df.shape, less the number of missing values for each particular column, if there are any.

Task 6

The 'count' summary statistic gives the number of non-missing values for a variable. Missing values are excluded from the count.

'mean' gives arithmetic average of a variable. It is calculated by summing all the non-missing values and dividing by the count of non-missing values.

'std' refers to standard deviation, which measures the spread of values around the mean of a variable. The higher the standard deviation, the more dispersed the values are from the mean.

'min' gives the smallest value in the variable.

'25%' refers to the first quartile, or 25th percentile. It is the value below which 25% of the data points fall when the data points are arranged in ascending order. 

'50%' refers to the median, or 50th percentile. It is the middle value of a variable when the data points are arranged in ascending order. 

'75%' refers to the third quartile, or 75th percentile when the data points are arranged in ascending order.

'max' gives the largest value in the variable.

Now suppose 3 patients (observations) have missing entries for age, date of admission, room number and diagnosis. Using _df.dropna()_ will enable us to remove these 3 patients from the dataset, ensuring that our analysis is based only on patients with full data as we wanted. On the other hand, had we used _del df['diagnosis']_, for example, we would have removed information about the diagnosis of all patients, including the other 97 who did not have missing values for diagnosis. This would significantly inhibit our analysis.

Part 2:

Given that the 'allergies' column is mostly empty, it is deemed not important to our analysis (which requires observations with full data as assumed above). 

Using _del df['allergies']_ will allow us to delete the allergies column while keeping all the other variables such as patient ID, name, age, etc. intact. Meanwhile, using _df.dropna()_ in this instance would mean losing the data entries for all the patients who did not have any known allergies, and who make up the majority of the observations. We would lose a lot of data as a result.

Applying _del df['col']_ first allows us to remove variables which may not be relevant to our analysis, and which may contain missing values.
